```python
temperature = 25         # use of if,elif, and else statement 

if temperature > 30:
    print("it's a hot day")
    print("Drink plenty of water")
elif temperature > 20: # (20, 30)
    print("it's a nice day")
elif temperature > 10: # (10, 20)
    print("it's a bit cold")
else:
    print("it's cold")
print('done')
```

```python
weight = float(input('weight: '))    # a weight unit conversion, it takes an input with its unit and convert it to the other unit 
unit = input('(K)g or (L)bs: ')
if unit.upper() == 'K':
    converted_weight = weight * 0.45
    print(converted_weight)
else:
    converted_weight = weight / 0.45
    print(converted_weight)
```
